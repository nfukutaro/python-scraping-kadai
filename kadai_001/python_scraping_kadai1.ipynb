```python
import requests
from bs4 import BeautifulSoup

#　記事を取得する対象のURL
url = 'https://news.yahoo.co.jp/articles/43ffc05e0311f51267a950121a095a84d8670132'

#  GETリクエストを送信し、HTMLを取得
response = requests.get(url)

#  HTMLを解析する
soup = BeautifulSoup(response.text,'html.parser')

#　記事本文の要素を取得する
#　CSSセレクタを利用して記事本文を取得する
article_element = soup.select_one('#uamods > div> div >p')

#  記事本文のテキストを取得
article_text = article_element.text

#  記事本文のテキストを表示
print(article_text)
```
